```python
df.drop(['Transmission', 'Power', 'Engine CC'],
        axis = 1,
        inplace = True)
df.columns
```

```python
df.columns = list(map(str,df.columns))
df.columns
```

```python
### 1. What is the mean price of every Manufacturer
```

```python
myPivot = pd.pivot_table(df,
                         index= ['Manufacturer'],
                         values= ['Price'],
                         aggfunc= ['mean'],
                        )
myPivot
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
myPivot.plot(figsize=(15, 10),legend = False, kind= 'bar')
plt.xticks(fontsize = 10, rotation = 90)
plt.title('Price', fontsize = 15)
plt.show()
```
