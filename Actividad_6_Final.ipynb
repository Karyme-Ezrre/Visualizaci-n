```python
### 1. What is the mean price of every Manufacturer
```

```python
myPivot = pd.pivot_table(df,
                         index= ['Manufacturer'],
                         values= ['Price'],
                         aggfunc= ['mean'],
                        )
myPivot
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
myPivot.plot(figsize=(15, 10),legend = False, kind= 'bar')
plt.xticks(fontsize = 10, rotation = 90)
plt.title('Price', fontsize = 15)
plt.show()
```

```python
z = pd.read_csv('CarBuyers.csv')
z.head()
```

```python
z.rename(columns = {'Male':'Male Buyer', 'Female': 'Female Buyer', 'Unknown' : 'Unknown Buyer',
       'Total': 'Total Buyers',},
          inplace = True
         )
```

```python
z.head()
```

```python
z.drop(['Model','Price','Transmission', 'Power', 'Engine CC','Fuel'],
        axis = 1,
        inplace = True)
z.columns
```

```python
z.head()
```

```python
z.set_index(['Manufacturer'], inplace = True)
```

```python
z.head()
```

```python
z.columns
```

```python
z.dtypes
```

```python
z = z.replace(',','', regex=True)
```

```python
z.to_numeric(z,['Male Buyer', 'Female Buyer', 'Unknown Buyer', 'Total Buyers'],errors = 'coerce')
```

```python
z.groupby(['Manufacturer']).agg({'Male Buyer' : 'sum', 'Female Buyer' : 'sum', 'Unknown Buyer' : 'sum',
       'Total Buyers' : 'sum'})
```
